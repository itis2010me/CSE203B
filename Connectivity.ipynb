```python
# create the 58x58 matrix
county_matrix = np.zeros((58, 58))

# iterate over the dictionary to populate the matrix
for i, county in enumerate(counties):
    for neighbor in adjacency_dict[county]:
        j = counties.index(neighbor)
        county_matrix[i][j] = 1
        print("County: ",county, " Neighbor: ",neighbor)

count = 0
for i in range(len(county_matrix)):
    for j in range(len(county_matrix[0])):
        if county_matrix[i][j] != 0:
            count += 1
    

print(count)
```
